```python
######################################################
## Clustering stocks using KMeans            #########
######################################################
import warnings
warnings.filterwarnings('ignore')

# You'll cluster companies using their daily stock price movements:
# the dollar difference between the closing and opening prices for each trading day. 
# You are given a NumPy array movements of daily price movements from 2010 to 2015 
# obtained from Yahoo! Finance, where each row corresponds to a company, 
# and each column corresponds to a trading day.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

movements = pd.read_csv("investment.csv")

display(movements.head(), '\n')

#a list companies of the company names is available.
companies = movements.iloc[:,0].tolist()
display(companies)

movements=np.array(movements)
print(movements[:5], '\n')
movements = np.delete(movements, 0, axis=1)
print(movements[:5])

```

```python
#################################################
# Features normalization to improve clustering
#################################################
# Some stocks are more expensive than others. 
# To account for this, include a Normalizer at the beginning of your pipeline. 
# The Normalizer will separately transform each company's stock price 
# to a relative scale before the clustering begins.

# While StandardScaler() standardizes features (such as the features of the fish data)
# by removing the mean (mean=0) and scaling to unit variance (variance=1), 
# Normalizer() rescales each sample - here, each company's stock price - independently of the other.

# Perform the necessary imports
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans

# Create a normalizer: normalizer
normalizer = Normalizer()

# Create a KMeans model with 10 clusters: kmeans
kmeans = KMeans(n_clusters = 10)

# Make a pipeline chaining normalizer and kmeans: pipeline
pipeline = make_pipeline(normalizer, kmeans)

# Fit pipeline to the daily price movements
pipeline.fit(movements)

# Now that your pipeline has been set up, 
# you can find out which stocks move together.

```

```python
########################################################
# which stocks move together
########################################################
# companies is the list of the company names.

# Predict the cluster labels: labels
labels = pipeline.predict(movements)

# Create a DataFrame aligning labels and companies: df
df = pd.DataFrame({'labels': labels, 'companies': companies})

# Display df sorted by cluster label
print(df.sort_values(by='labels'))

# You can see which companies are together in each cluster.
# Take a look at the clusters. Are you surprised by any of the results? 
```
